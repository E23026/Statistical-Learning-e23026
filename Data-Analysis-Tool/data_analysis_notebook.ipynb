# Data Analysis Tool

Data Analysis Tool is an object-oriented Python package developed to support data sanitization and exploratory data analysis. It offers capabilities for data inspection, missing value treatment, duplicate removal, outlier handling, feature transformation, correlation analysis, and interactive visualization, making it a comprehensive solution for preprocessing and understanding datasets.


````python
import pandas as pd

from data_analysis import (
    DataInspector,
    PlottingMethods
)
````

## Quick Start Example

````python
inspector = DataInspector()
````

## Data Import

Upload a CSV file or load sample Titanic dataset.

````python
inspector.upload_data()
````

````python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
inspector.df = pd.read_csv(url)
````

## Initial Data Inspection and Cleaning

````python
inspector.get_summary()
````

````python
inspector.column_details()
````

````python
inspector.show_missing_data()
````

````python
inspector.handle_missing_values(
    strategy="median"
)
````

````python
inspector.remove_duplicates()
````

## Numerical Visualization

````python
column_names = [
    "Age",
    "Fare"
]

inspector.plot_numerical(
    column_names=column_names
)
````

## Outlier Detection

````python
inspector.handle_outliers(
    columns=["Age", "Fare"],
    find_and_delete=True
)
````

## Categorical Analysis

````python
inspector.plot_categorical(
    column_names=[
        "Sex",
        "Embarked"
    ]
)
````

## Relationship Analysis

````python
# numeric vs numeric
inspector.plot_relationship(
    "Age",
    "Fare"
)
````

````python
# categorical vs numerical
inspector.plot_relationship(
    "Sex",
    "Fare"
)
````

## Correlation Analysis

````python
inspector.plot_numerical_correlation()
````

````python
inspector.plot_categorical_correlation()
````

````python
inspector.plot_all_associations_heatmap()
````

## Feature Engineering and Normalization

````python
normalized_numeric = (
    inspector.extract_normalized_numeric_data(
        method="robust"
    )
)

normalized_numeric.head()
````

````python
encoded_cat = (
    inspector.extract_normalized_categorical_data(
        method="onehot"
    )
)

encoded_cat.head()
````

````python
final_df = (
    inspector.create_normalized_data_df()
)

final_df.head()
````

## Custom Plotting Methods

````python
PLT = PlottingMethods()
````

````python
PLT.plot_pie_chart(
    names="Sex",
    values="Fare",
    data=inspector.df
)
````

````python
PLT.plot_bar_chart(
    x="Sex",
    y="Fare",
    data=inspector.df
)
````

````python
PLT.plot_histogram(
    x="Age",
    data=inspector.df
)
````
