# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


1. In this model, the probability of an event $A$ is proportional to its area.  Sample Space ($\Omega$): The disk of radius 10,

Ω={(x,y)∈R$^2$:x$^2$+y$^2$≤10$^2$}

Total Area: $\pi(10)^2 = 100\pi$.

Area of $A$: (Area of disk radius 10) - (Area of disk radius 9) $= \pi(10)^2 - \pi(9)^2 = 100\pi - 81\pi = 19\pi$.

$$P(A) = \frac{19\pi}{100\pi} = \frac{19}{100} = 0.19$$

2.  Variables: Radius $r$ is chosen uniformly from $[0, 10]$, and angle $\theta$ is chosen uniformly from $[0, 2\pi)$.  

Independence: Since the angle is independent and uniform, it does not affect the probability of falling within a certain distance from the center. We only need to consider the measure of the radius $r$.  

Total "Length" of Radius: $10 - 0 = 10$.  

Favorable "Length" for $A$: The event $A$ occurs if $9 \le r \le 10$, which has a length of $10 - 9 = 1$.  

Calculation:$$P(A) = \frac{\text{Favorable Length}}{\text{Total Length}} = \frac{1}{10} = 0.10$$

3. The difference arises because you are using two different probability measures on the same geometric set $\Omega$.

Area-Uniform Model: Points near the boundary are more likely because the outer rings of a disk contain significantly more area than the inner rings. This is the standard model for physical "random scattering," such as throwing darts at a board.  

Radius-Uniform Model: Every "ring" of the same width is treated as equally likely, regardless of how much area it actually occupies. This corresponds to an artificial two-step sampling rule rather than natural geometric randomness.

1. The Probability Space $(\Omega, \mathcal{F}, P$)

In measure theory, a probability space consists of the sample space, the set of all "measurable" events, and the probability measure.

Sample Space ($\Omega$): The set of all possible outcomes.
$$\Omega = \{G_1, G_2, D_1, D_2\}$$$\sigma$-algebra ($\mathcal{F}$): Since the problem states $\mathcal{F} = \mathcal{P}(\Omega)$, it is the power set containing all $2^4 = 16$ possible subsets of $\Omega$.$$\mathcal{F} = \{ \emptyset, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1, G_2\}, \{G_1, D_1\}, \dots, \Omega \}$$
Probability Measure ($P$): The assignment of weights to the atoms of the space.$$P(\{G_1\})=0.50, \quad P(\{G_2\})=0.20, \quad P(\{D_1\})=0.10, \quad P(\{D_2\})=0.20$$
2. Compute the $\sigma$-algebra generated by $X$, $\sigma(X)$The $\sigma$-algebra $\sigma(X)$ represents the information revealed by $X$. It is the collection of all "pull-backs" of sets from the range of $X$.

$X$ takes the value $0$ for the set $\{G_1, G_2\}$.
$X$ takes the value $1$ for the set $\{D_1, D_2\}$.
$$\sigma(X) = \{ \emptyset, \Omega, \{G_1, G_2\}, \{D_1, D_2\} \}$$
3. Compute the $\sigma$-algebra generated by $Y$, $\sigma(Y)$
Since $Y$ assigns a unique value to every single outcome in $\Omega$ (1, 2, 3, and 4 respectively), every individual outcome is "measurable" or "distinguishable" by $Y$.

$$\sigma(Y) = \mathcal{P}(\Omega) \text{ (The full power set)}$$
This contains all 16 possible subsets of $\Omega$.

4. Show $\sigma(X) \subseteq \sigma(Y)$ and Information Comparison

Proof of Inclusion: Look at the elements of $\sigma(X)$. $\emptyset$ and $\Omega$ are in $\sigma(Y)$ by definition. The set $\{G_1, G_2\}$ is in $\sigma(Y)$ because it is the union of atoms $\{G_1\}$ and $\{G_2\}$, which are both in $\sigma(Y)$. The same applies to $\{D_1, D_2\}$. Thus, every set in $\sigma(X)$ is also in $\sigma(Y)$.

Information Interpretation: $Y$ contains more information because it is a "finer" partition of the sample space. If you know $Y$, you know exactly which machine produced the item and its quality. If you only know $X$, you know the quality but have "lost" the information regarding which machine produced it. In mathematical terms, $X$ is $Y$-measurable (there exists a function $f$ such that $X = f(Y)$).

The distribution of $X$ is found by summing the probabilities of the outcomes that result in each value of $X$.
$P(X=0) = P(\{G_1, G_2\}) = P(G_1) + P(G_2) = 0.50 + 0.20 = \mathbf{0.70}$

$P(X=1) = P(\{D_1, D_2\}) = P(D_1) + P(D_2) = 0.10 + 0.20 = \mathbf{0.30}$
