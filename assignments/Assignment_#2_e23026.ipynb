#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



# Answer

1. Construct a suitable probability space $(\Omega, \mathfrak{F}, P)$

In measure theory, a probability space consists of a sample space $\Omega$, a $\sigma$-algebra $\mathfrak{F}$, and a probability measure $P$.  
Sample Space ($\Omega$): The set of all possible outcomes. Here, $\Omega = \{0, 1, 2, 3\}$.  
$\sigma$-algebra ($\mathfrak{F}$): Since the sample space is finite, we use the Power Set $\mathfrak{F} = \mathfrak{P}(\Omega)$, which contains all $2^4 = 16$ possible subsets of $\Omega$. This is the most granular $\sigma$-algebra possible for a discrete space.  
Probability Measure ($P$): A function $P: \mathfrak{F} \rightarrow [0, 1]$ defined by assigning values to singletons (individual outcomes) and using additivity for other events:  $P(\{0\}) = 0.50$   $P(\{1\}) = 0.30$   $P(\{2\}) = 0.15$   $P(\{3\}) = 0.05$   
For any $A \in \mathfrak{F}$, $P(A) = \sum_{\omega \in A} P(\{\omega\})$.  

2. Define $X$ as a random variable

A random variable is a measurable function $X: \Omega \rightarrow \mathbb{R}$. In this model, we define the identity mapping:
  $$X(\omega) = \omega, \quad \forall \omega \in \Omega$$.
This map is measurable because for any Borel set $B \subseteq \mathbb{R}$, the pre-image $X^{-1}(B)$ is a subset of $\Omega$ and therefore is an element of our $\sigma$-algebra $\mathfrak{F}$.  

3. Probability Mass Function (PMF)

The PMF $p_X(x)$ is defined as $P(X = x)$:
  $$p_X(x) = \begin{cases} 0.50 & \text{if } x = 0 \\ 0.30 & \text{if } x = 1 \\ 0.15 & \text{if } x = 2 \\ 0.05 & \text{if } x = 3 \\ 0 & \text{otherwise} \end{cases}$$.  
  
  4. Compute the Expectation $E[X]$
  
  The expectation for a discrete random variable is the weighted average of its values:
  $$E[X] = \sum_{x} x \cdot p_X(x)$$.
  $$E[X] = 0(0.50) + 1(0.30) + 2(0.15) + 3(0.05) = 0.75$$.  
  
  5. Compute the Variance $Var(X)$
  
  To find the variance, we first compute the second moment $E[X^2]$:
  $$E[X^2] = \sum_{x} x^2 \cdot p_X(x) = 0^2(0.50) + 1^2(0.30) + 2^2(0.15) + 3^2(0.05) = 1.35$$.
Now, use the formula $Var(X) = E[X^2] - (E[X])^2$:
  $$Var(X) = 1.35 - (0.75)^2 = 1.35 - 0.5625 = 0.7875$$.  
  
  6. Find $P(X \geq 1)$ and $P(X \geq 2)$
  
  $P(X \geq 1) = P(X=1) + P(X=2) + P(X=3) = 0.30 + 0.15 + 0.05 = 0.50$.  $P(X \geq 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = 0.20$.  
  
  7. Practical Meaning
  
  Expectation ($E[X] = 0.75$): This represents the long-run average number of emergency transformer failures per day. Over many days, the district can expect approximately 75 failures for every 100 days.  
  
  Variance ($Var(X) = 0.7875$): This measures the "spread" or uncertainty in the daily failure rate. A higher variance would indicate that the number of daily failures fluctuates significantly from day to day, whereas a low variance would suggest the number of failures is consistently close to the mean.

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


# Answer
1. Construct a suitable probability space $(\Omega, \mathfrak{F}, P)$

For a continuous random variable modeling a proportion on the interval $[0, 1]$, the probability space is constructed as follows:

Sample Space ($\Omega$): The set of all possible outcomes for the proportion. Here, $\Omega = [0, 1]$.  

$\sigma$-algebra ($\mathfrak{F}$): Since the sample space is an uncountable interval, we use the Borel $\sigma$-algebra $\mathfrak{B}([0, 1])$, which is the smallest $\sigma$-algebra containing all open intervals in $[0, 1]$.  

Probability Measure ($P$): A measure defined for any $A \in \mathfrak{F}$ by the integral of the density function:$$P(A) = \int_A f(x) \, dx = \int_A \frac{1}{B(2,18)} x^{1}(1-x)^{17} \, dx$$.  

2. Define $X$ as a continuous random variable

A random variable is a measurable function $X: \Omega \rightarrow \mathbb{R}$. On this space, we define $X$ as the identity mapping:
  $$X(\omega) = \omega, \quad \forall \omega \in [0, 1]$$This mapping is measurable because for any Borel set $B \subseteq \mathbb{R}$, its pre-image $X^{-1}(B) \cap [0,1]$ is a Borel subset of $[0, 1]$ and thus an element of $\mathfrak{F}$.
  
  3. Probability Density Function (PDF)
  
  The probability density function is given by:$$f(x) = \begin{cases} \frac{1}{B(2,18)} x^{1}(1-x)^{17} & \text{for } 0 < x < 1 \\ 0 & \text{otherwise} \end{cases}$$where $B(\alpha, \beta)$ is the Beta function defined as $B(\alpha, \beta) = \int_0^1 t^{\alpha-1}(1-t)^{\beta-1} dt$. For $\alpha=2, \beta=18$, $B(2,18) = \frac{\Gamma(2)\Gamma(18)}{\Gamma(20)} = \frac{1! \cdot 17!}{19!} = \frac{1}{19 \cdot 18} = \frac{1}{342}$. Thus, $f(x) = 342x(1-x)^{17}$ for $x \in (0,1)$.
  
  4. Verify that $f$ is a valid PDF
  
  A valid PDF must satisfy two conditions:  
  Non-negativity: $f(x) \geq 0$ for all $x$. Since $x \in [0,1]$, both $x$ and $(1-x)$ are non-negative, so $f(x) \geq 0$.
  Normalization: The total integral must be 1.$$\int_{-\infty}^{\infty} f(x) dx = \int_0^1 \frac{x^{2-1}(1-x)^{18-1}}{B(2,18)} dx = \frac{1}{B(2,18)} \int_0^1 x^{2-1}(1-x)^{18-1} dx = \frac{B(2,18)}{B(2,18)} = 1$$.
  
  5. Compute the Expectation $E[X]$
  
  For a Beta distribution with parameters $\alpha$ and $\beta$, the mean is given by:$$E[X] = \frac{\alpha}{\alpha + \beta} = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$.
  
  6. Compute the Variance $Var(X)$
  
  The formula for the variance of a Beta distribution is:$$Var(X) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)} = \frac{2 \cdot 18}{(20)^2(21)} = \frac{36}{400 \cdot 21} = \frac{36}{8400} = \frac{3}{700} \approx 0.004286$$.
  
  7. Find $P(X \leq 0.10)$
  
  This is the integral of the density from 0 to 0.10:$$P(X \leq 0.10) = \int_0^{0.10} \frac{1}{B(2,18)} x(1-x)^{17} dx$$.
  
  8. Find $P(0.05 \leq X \leq 0.15)$
  
  This is the integral over the specified interval:$$P(0.05 \leq X \leq 0.15) = \int_{0.05}^{0.15} \frac{1}{B(2,18)} x(1-x)^{17} dx$$.
  
  9. Why Beta distribution is a reasonable model
  
  The Beta distribution is a standard choice for modeling proportions or probabilities because its domain is strictly bounded between 0 and 1. It is highly flexible; by adjusting $\alpha$ and $\beta$, it can model various shapes of uncertainty, such as the likelihood that the defective rate is typically low but has some variability.
  
  10. Interpret Expectation and Variance
  
  Expectation ($E[X] = 0.10$): This represents the long-run average defective rate. On average, 10% of items produced are expected to be defective.
  
  Variance ($Var(X) \approx 0.0043$): This measures the consistency of production quality. A low variance indicates that the daily defective proportion is usually very close to 10%, whereas a high variance would suggest that quality fluctuates significantly from day to day.

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


# Answer

1. Support of $X$

The support of a random variable is the set of values where the density is non-zero. For this model:$$\text{Support}(X) = [4, 10]$$

2. Probability Density Function $f_X(x)$

The PDF of a Triangular distribution is defined piecewise:$$f_X(x) = \begin{cases} \frac{2(x-a)}{(b-a)(c-a)} & \text{for } a \leq x < c \\ \frac{2(b-x)}{(b-a)(b-c)} & \text{for } c \leq x \leq b \\ 0 & \text{otherwise} \end{cases}$$Plugging in $a=4, b=10, c=6$:$$f_X(x) = \begin{cases} \frac{2(x-4)}{(6)(2)} = \frac{x-4}{6} & \text{for } 4 \leq x < 6 \\ \frac{2(10-x)}{(6)(4)} = \frac{10-x}{12} & \text{for } 6 \leq x \leq 10 \\ 0 & \text{otherwise} \end{cases}$$

3. Verify that $f_X$ integrates to 1

We integrate over the two pieces of the support:$$\int_4^6 \frac{x-4}{6} dx + \int_6^{10} \frac{10-x}{12} dx$$

First part: $\left[ \frac{(x-4)^2}{12} \right]_4^6 = \frac{4}{12} - 0 = \frac{1}{3}$

Second part: $\left[ -\frac{(10-x)^2}{24} \right]_6^{10} = 0 - (-\frac{16}{24}) = \frac{2}{3}$

Total integral: $\frac{1}{3} + \frac{2}{3} = 1$.

4. Expectation $E[X]$

The mean of a Triangular distribution is $\frac{a+b+c}{3}$:$$E[X] = \frac{4 + 10 + 6}{3} = \frac{20}{3} \approx 6.67 \text{ hours.}$$

5. Variance $Var(X)$

The formula for variance is $\frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}$:$$Var(X) = \frac{4^2 + 10^2 + 6^2 - (40) - (24) - (60)}{18} = \frac{16 + 100 + 36 - 124}{18} = \frac{28}{18} = \frac{14}{9} \approx 1.56.$$

6. Interpretation of $a, b, c$

$a=4$: The "optimistic" time; the absolute minimum time required if everything goes perfectly.

$b=10$: The "pessimistic" time; the maximum time allowed for unexpected delays.

$c=6$: The "most likely" time; the peak of the distribution where the probability density is highest.

7. Why a Triangular distribution is reasonable

It is a "subjective" distribution used when actual historical data is scarce. Experts can usually estimate a minimum, maximum, and a most likely value more easily than they can estimate parameters like standard deviation for a Normal distribution.

8. Probability $P(X \leq 5)$

Since $5$ is in the first interval $[4, 6)$:$$P(X \leq 5) = \int_4^5 \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_4^5 = \frac{1}{12} \approx 0.0833.$$

9. Probability $P(5 \leq X \leq 8)$

We calculate $P(X \leq 8)$ and subtract $P(X \leq 5)$. Since $8$ is in the second interval:$$P(X > 8) = \int_8^{10} \frac{10-x}{12} dx = \left[ -\frac{(10-x)^2}{24} \right]_8^{10} = 0 - (-\frac{4}{24}) = \frac{1}{6}.$$$$P(X \leq 8) = 1 - P(X > 8) = 1 - \frac{1}{6} = \frac{5}{6}.$$Therefore, $P(5 \leq X \leq 8) = P(X \leq 8) - P(X \leq 5) = \frac{5}{6} - \frac{1}{12} = \frac{10-1}{12} = \frac{9}{12} = 0.75$.

10. Use in Planning and Sensitivity Analysis

This model is used in Monte Carlo simulations for project schedules (like PERT). By identifying the variance, planners can perform sensitivity analysis to see how much the total project timeline changes if this specific task (generator installation) encounters a "worst-case" scenario ($b=10$). It helps in setting realistic deadlines and buffer times.

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is given by

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


# Answer

1. Compute the shape parameters $\alpha$ and $\beta$

The PERT distribution uses the "four-times-most-likely" rule to determine its shape:$\alpha = 1 + 4\left(\frac{c-a}{b-a}\right) = 1 + 4\left(\frac{11-8}{20-8}\right) = 1 + 4\left(\frac{3}{12}\right) = 1 + 1 = 2$$\beta = 1 + 4\left(\frac{b-c}{b-a}\right) = 1 + 4\left(\frac{20-11}{20-8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = 4$

2. Specify the support of $X$

The support is the interval of possible values for the task duration:$$\text{Support}(X) = [8, 20]$$

3. Probability Density Function $f_X(x)$

Using $\alpha=2, \beta=4, a=8, b=20$:$$f_X(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2,4)(20-8)^{2+4-1}} = \frac{(x-8)(20-x)^3}{B(2,4) \cdot 12^5}, \quad \text{for } 8 \leq x \leq 20$$(Note: $B(2,4) = \frac{\Gamma(2)\Gamma(4)}{\Gamma(6)} = \frac{1! \cdot 3!}{5!} = \frac{6}{120} = \frac{1}{20}$)

4. State the probability space $(\Omega, \mathfrak{F}, P)$

Sample Space ($\Omega$): The interval $[8, 20]$.

$\sigma$-algebra ($\mathfrak{F}$): The Borel $\sigma$-algebra $\mathfrak{B}([8, 20])$.

Probability Measure ($P$): Defined for any $A \in \mathfrak{F}$ by $P(A) = \int_A f_X(x) dx$.

5. Define $X$ as a random variable

$X$ is the measurable identity function $X: \Omega \to \mathbb{R}$ where $X(\omega) = \omega$ for all $\omega \in [8, 20]$.

6. Show that $E[X] = \frac{a+4c+b}{6}$

For a scaled Beta distribution on $[a, b]$, the mean is $a + (b-a)\frac{\alpha}{\alpha+\beta}$.Substituting our $\alpha$ and $\beta$ formulas:$E[X] = a + (b-a)\frac{1 + 4\frac{c-a}{b-a}}{6} = a + \frac{(b-a) + 4(c-a)}{6} = \frac{6a + b - a + 4c - 4a}{6} = \frac{a + 4c + b}{6}$.Plugging in values: $E[X] = \frac{8 + 4(11) + 20}{6} = \frac{72}{6} = 12$ days.

7. Compute the Variance $Var(X)$

Using the provided approximation:$Var(X) = \frac{(12-8)(20-12)}{7} = \frac{4 \cdot 8}{7} = \frac{32}{7} \approx 4.57$.

8. Compute the approximate standard deviation $\sigma$

$\sigma = \sqrt{Var(X)} = \sqrt{\frac{32}{7}} \approx 2.14$ days.

9. PERT vs. Triangular Distribution

The PERT distribution is generally more appropriate than the Triangular distribution because it is smoother and places less probability mass at the extremes ($a$ and $b$). While the Triangular distribution mean for these values would be $\frac{8+11+20}{3} = 13$, the PERT mean is $12$. PERT emphasizes the "most likely" value ($c$) more strongly, which often aligns better with expert intuition in software projects where extreme delays are possible but the peak of the probability is concentrated.

10. Practical Interpretations

$a=8$: The best-case scenario (no bugs, perfect integration).

$b=20$: The worst-case scenario (major compatibility issues).

$c=11$: The most frequent result expected if the task were repeated many times.

$E[X]=12$: The "risk-adjusted" average; this is the value a manager should use in a project schedule to account for uncertainty.

$\sigma=2.14$: The level of risk; a larger $\sigma$ indicates higher uncertainty and suggests the need for a larger "contingency buffer" in the project plan.

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

# Answer

1. Determine the value of the maximum time $b$

In a PERT distribution, the expectation $E[X]$ is given by the weighted average formula:$$E[X] = \frac{a + 4c + b}{6}$$The problem states that the expectation must match the most likely time (the mode), meaning $E[X] = c$. We can set up the following equation:$$\frac{a + 4c + b}{6} = c$$Now, solve for $b$:

Multiply both sides by 6: $a + 4c + b = 6c$

Subtract $a$ and $4c$ from both sides: $b = 6c - 4c - a$

Simplify: $b = 2c - a$Plugging in the given values $a = 5$ and $c = 11$:$$b = 2(11) - 5$$$$b = 22 - 5 = 17$$Answer: The maximum time $b$ must be 17 days.

2. What happens to the variance?

In this specific scenario where $E[X] = c$, the distribution becomes symmetric.Let's look at the shape parameters $\alpha$ and $\beta$ when $b = 2c - a$:$\alpha = 1 + 4\left(\frac{c-a}{b-a}\right) = 1 + 4\left(\frac{c-a}{(2c-a)-a}\right) = 1 + 4\left(\frac{c-a}{2c-2a}\right) = 1 + 4\left(\frac{1}{2}\right) = 3$$\beta = 1 + 4\left(\frac{b-c}{b-a}\right) = 1 + 4\left(\frac{(2c-a)-c}{2c-2a}\right) = 1 + 4\left(\frac{c-a}{2c-2a}\right) = 1 + 4\left(\frac{1}{2}\right) = 3$Since $\alpha = \beta = 3$, the probability density function is perfectly balanced around the mean.

Effect on Variance:When $E[X] = c$, the "skewness" of the distribution is zero. In the context of project management, this means the risk of a "late finish" (the right tail) is perfectly balanced by the possibility of an "early finish" (the left tail).Using the approximation formula $Var(X) = \frac{(E[X]-a)(b-E[X])}{7}$:$$Var(X) = \frac{(11-5)(17-11)}{7} = \frac{6 \times 6}{7} = \frac{36}{7} \approx 5.14$$If the mode $c$ were shifted closer to the minimum $a$ (making $E[X] > c$), the variance would likely increase because the "tail" towards the maximum $b$ would become much longer, representing higher uncertainty and "downside risk." By forcing $E[X] = c$, we have created a more stable, predictable model where the most frequent outcome is also the average outcome.
