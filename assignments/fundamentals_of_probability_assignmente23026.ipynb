# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


1. In this model, the probability of an event $A$ is proportional to its area.  Sample Space ($\Omega$): The disk of radius 10,

Ω={(x,y)∈R$^2$:x$^2$+y$^2$≤10$^2$}

Total Area: $\pi(10)^2 = 100\pi$.

Area of $A$: (Area of disk radius 10) - (Area of disk radius 9) $= \pi(10)^2 - \pi(9)^2 = 100\pi - 81\pi = 19\pi$.

$$P(A) = \frac{19\pi}{100\pi} = \frac{19}{100} = 0.19$$

2.  Variables: Radius $r$ is chosen uniformly from $[0, 10]$, and angle $\theta$ is chosen uniformly from $[0, 2\pi)$.  

Independence: Since the angle is independent and uniform, it does not affect the probability of falling within a certain distance from the center. We only need to consider the measure of the radius $r$.  

Total "Length" of Radius: $10 - 0 = 10$.  

Favorable "Length" for $A$: The event $A$ occurs if $9 \le r \le 10$, which has a length of $10 - 9 = 1$.  

Calculation:$$P(A) = \frac{\text{Favorable Length}}{\text{Total Length}} = \frac{1}{10} = 0.10$$

3. The difference arises because you are using two different probability measures on the same geometric set $\Omega$.

Area-Uniform Model: Points near the boundary are more likely because the outer rings of a disk contain significantly more area than the inner rings. This is the standard model for physical "random scattering," such as throwing darts at a board.  

Radius-Uniform Model: Every "ring" of the same width is treated as equally likely, regardless of how much area it actually occupies. This corresponds to an artificial two-step sampling rule rather than natural geometric randomness.

5. Marginal Density of $X$To find the marginal density $f_X(t)$, we integrate the joint density over all possible values of $p$:$$f_X(t) = \int_{-\infty}^{\infty} f_{T,P}(t, p) \, dp = \int_{-\infty}^{\infty} \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right) \, dp$$Factoring out the terms involving $t$:$$f_X(t) = \frac{1}{\sqrt{2\pi}} e^{-t^2/2} \left[ \int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi}(2)} e^{-p^2/(2 \cdot 2^2)} \, dp \right]$$The term in the bracket is the integral of a $N(0, 4)$ density, which equals 1. Thus:$f_X(t) = \frac{1}{\sqrt{2\pi}} e^{-t^2/2}$, which is the density of $N(0,1)$.

6. Marginal Density of $Y$By definition, $Y = (T, P)$. The "marginal" density of the vector $Y$ is simply the joint density of its components:$f_Y(t, p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$.

7. Probability Computations$P(X \le 0)$: Since $X \sim N(0,1)$ is symmetric around 0, $P(X \le 0) = 0.5$.$P(X > 1)$: Using the standard normal distribution table ($Z$-table), $P(X > 1) = 1 - \Phi(1) \approx 1 - 0.8413 = 0.1587$.$P(Y \in (-\infty, 0] \times \mathbb{R})$: This is the probability that $T \le 0$ regardless of $P$. This is equivalent to $P(X \le 0) = 0.5$.$P(Y \in [-1, 1] \times [-2, 2])$: Due to independence:$$P(-1 \le T \le 1) \cdot P(-2 \le P \le 2)$$For $T \sim N(0,1)$: $P(-1 \le T \le 1) \approx 0.6827$.For $P \sim N(0,4)$, the standardized variable is $Z = P/2 \sim N(0,1)$. Thus, $P(-2 \le P \le 2) = P(-1 \le Z \le 1) \approx 0.6827$.Total probability $\approx 0.6827^2 \approx 0.4661$.

8. Conditional Distribution of $X$ Given $Y=(t,p)$In measure theory, conditioning on the "full information" $\sigma(Y)$ where $Y$ is the identity map means the outcome is fully known.Since $X$ is a function of $Y$ (specifically, $X$ is the projection of $Y$ onto the first coordinate), $X$ is $\sigma(Y)$-measurable.Therefore, given $Y = (t, p)$, $X$ is no longer random; it is exactly $t$.The conditional distribution is a Dirac delta distribution centered at $t$, denoted $\delta_t$.Formally, $E[X | Y] = X$, because conditioning on full information leaves a measurable variable unchanged.

9. Conditional Distribution of $Y$ given $X=t$Since $T$ and $P$ are independent, the joint density factors as $f_{T,P}(t,p) = f_T(t)f_P(p)$. The conditional density is given by:$$f_{Y|X}(y|t) = f_{(T,P)|T}(t,p|t)$$Because $T$ is fixed at $t$, the first component is deterministic ($t$) and the second component $P$ remains independent of $T$. Thus, the conditional distribution of $Y$ given $X=t$ is the distribution of the random vector $(t, P)$, where $P \sim N(0, 4)$.Formally: The conditional distribution is a probability measure on $\mathbb{R}^2$ concentrated on the line $T=t$, specifically $\delta_t \otimes N(0, 4)$.

10. Information Comparison: $X$ vs. $Y$Conditioning on $X$: When we condition on $X$, we know the exact value of the reactor temperature deviation. However, we have zero information about the pressure deviation $P$; our uncertainty about $P$ remains exactly its prior distribution, $N(0, 4)$.Conditioning on $Y$: When we condition on $Y$, we know both the temperature $T$ and the pressure $P$. The "additional information" is the specific value of $P$, which collapses the uncertainty of the pressure deviation to a single point.

11. $X$ as a Measurable Function of $Y$Proof: By definition, $Y(t,p) = (t,p)$ and $\pi_1(t,p) = t$. Therefore, $(\pi_1 \circ Y)(t,p) = \pi_1(t,p) = t = X(t,p)$.Measurability: Since $\pi_1: \mathbb{R}^2 \to \mathbb{R}$ is a continuous mapping (a projection), it is Borel-measurable. Because $X$ is the composition of a measurable function $\pi_1$ and the random variable $Y$, $X$ is $\sigma(Y)$-measurable. Intuitively, if you know the vector $(t,p)$, you automatically know $t$.

13. Differential Entropy $h(X)$For a normal distribution $N(\mu, \sigma^2)$, the differential entropy is $h(X) = \frac{1}{2} \ln(2\pi e \sigma^2)$.Given $X \sim N(0, 1)$:$h(X) = \frac{1}{2} \ln(2\pi e) \approx 1.4189$ nats.

14. Differential Entropy $h(Y)$For a multivariate normal $N(\mu, \Sigma)$, $h(Y) = \frac{1}{2} \ln((2\pi e)^k |\Sigma|)$, where $k=2$ and $|\Sigma|$ is the determinant of the covariance matrix.$\Sigma = \begin{pmatrix} 1 & 0 \\ 0 & 4 \end{pmatrix}$, so $|\Sigma| = 4$.$h(Y) = \frac{1}{2} \ln((2\pi e)^2 \cdot 4) = \ln(2\pi e) + \ln(2) \approx 2.112 + 0.693 = 2.805$ nats.(Note: Due to independence, $h(Y) = h(T) + h(P)$).

15. Conditional Differential Entropy $h(Y|X)$$h(Y|X) = h(T,P|T)$. Since $T$ is given, its entropy is zero, and we are left with the entropy of $P$.$h(Y|X) = h(P) = \frac{1}{2} \ln(2\pi e \cdot 4) \approx 2.112$ nats.Physical Interpretation: This represents the average amount of "uncertainty" or "information" remaining in the system sensors when the temperature deviation is already known. It measures the residual noise in the pressure sensor.

16. The Case of $h(X|Y)$Result: In this model, $h(X|Y) = -\infty$.Why: Because $X$ is a deterministic function of $Y$ ($X = \pi_1(Y)$), there is no uncertainty in $X$ once $Y$ is known. In the continuous case, the entropy of a certain (deterministic) value is $-\infty$ (unlike the discrete case where it is $0$).Difference: In discrete probability, if $X=f(Y)$, then $H(X|Y) = 0$. In continuous probability, conditioning a variable on itself or a variable that determines it results in a distribution with zero variance (a Dirac delta), which has a differential entropy of $-\infty$. This reflects the infinite "information density" required to specify a real number with perfect precision.
