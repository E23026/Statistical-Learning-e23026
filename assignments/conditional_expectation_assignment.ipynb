#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

In the context of prediction, the Mean Square Error (MSE) measures the average squared difference between the actual outcome ($X$) and our best estimate based on available information ($E[X|Y]$).In measure theory, this is often referred to as the Conditional Variance or the "residual uncertainty" left in $X$ after observing $Y$.

1. The MSE Formula

Since our sample space is discrete, the MSE is calculated as the expected value of the squared residuals:$$MSE = E\left[ (X - E[X|Y])^2 \right] = \sum_{\omega \in \Omega} (X(\omega) - E[X|Y](\omega))^2 \cdot P(\{\omega\})$$

2. Step-by-Step Calculation

We calculate the squared error for each outcome $\omega$:| Outcome | $X(\omega)$ | $E[X|Y](\omega)$ | Error $(X - \hat{X})$ | Squared Error | $P(\{\omega\})$ | Contribution || :--- | :--- | :--- | :--- | :--- | :--- | :--- || $\omega_1$ (Low/Legit) | 0 | $1/6$ | $-1/6$ | $1/36$ | $0.50$ | $0.50 \times \frac{1}{36} = \frac{1}{72}$ || $\omega_2$ (Low/Fraud) | 1 | $1/6$ | $5/6$ | $25/36$ | $0.10$ | $0.10 \times \frac{25}{36} = \frac{5}{72}$ || $\omega_3$ (High/Legit) | 0 | $1/2$ | $-1/2$ | $1/4$ | $0.20$ | $0.20 \times \frac{1}{4} = \frac{1}{20}$ || $\omega_4$ (High/Fraud) | 1 | $1/2$ | $1/2$ | $1/4$ | $0.20$ | $0.20 \times \frac{1}{4} = \frac{1}{20}$ |

3. Summing the Results

Now, we sum the contributions from all outcomes:$$MSE = \left( \frac{1}{72} + \frac{5}{72} \right) + \left( \frac{1}{20} + \frac{1}{20} \right)$$$$MSE = \frac{6}{72} + \frac{2}{20}$$Simplify the fractions:$\frac{6}{72} = \frac{1}{12}$$\frac{2}{20} = \frac{1}{10}$Find a common denominator (60):$$MSE = \frac{5}{60} + \frac{6}{60} = \mathbf{\frac{11}{60}}$$Numerical Value: $MSE \approx \mathbf{0.1833}$

4. Physical Interpretation

Total Variance vs. MSE: Before looking at the customer's risk class ($Y$), the total uncertainty in fraud ($X$) was its variance, $Var(X)$. By observing $Y$, we reduced that uncertainty. The $11/60$ represents the remaining error that the bank cannot eliminate simply by knowing the risk class.

The MMSE Property: A fundamental theorem in probability states that $E[X|Y]$ is the "best" possible predictor because it minimizes the MSE. Any other $\sigma(Y)$-measurable function (any other rule the bank could invent based on risk class) would result in an MSE strictly greater than $11/60$.

If the bank ignores the customer risk class $Y$ and uses the global average fraud rate ($E[X] = 0.30$) as its prediction for every single transaction, the error will increase.

By ignoring $Y$, the bank is effectively moving from a "conditional" prediction to a "marginal" prediction. The error of this global prediction is simply the Variance of $X$.

1. Calculation of the "Global" MSE ($Var(X)$)

The Mean Square Error for a constant estimate $c = E[X]$ is defined as:$$MSE_{global} = E[(X - E[X])^2] = Var(X)$$First, we calculate $E[X^2]$:$$E[X^2] = (0^2 \cdot 0.50) + (1^2 \cdot 0.10) + (0^2 \cdot 0.20) + (1^2 \cdot 0.20)$$$$E[X^2] = 0 + 0.10 + 0 + 0.20 = 0.30$$Now, using the formula $Var(X) = E[X^2] - (E[X])^2$:$$Var(X) = 0.30 - (0.30)^2 = 0.30 - 0.09 = \mathbf{0.21}$$

Comparison:

With information $Y$: $MSE \approx 0.1833$ (or $11/60$)

Without information $Y$: $MSE = 0.2100$ (or $12.6/60$)The error increases by approximately 14.5% when the bank ignores the risk class.

2. Relationship to the Law of Total Variance

The relationship between these two error values is governed by the Law of Total Variance, which decomposes the total uncertainty in $X$ into two distinct parts:$$Var(X) = E[Var(X|Y)] + Var(E[X|Y])$$Here is how each term corresponds to our fraud detection problem:

A. The "Unexplained" Error ($E[Var(X|Y)]$)

This is the MSE we calculated in the previous step ($\frac{11}{60} \approx 0.1833$).

Meaning: It represents the "within-group" variance. Even if you know a customer is high-risk, there is still uncertainty about whether this specific transaction is fraud. This is the noise that information $Y$ cannot filter out.

B. The "Explained" Variance ($Var(E[X|Y])$)

This is the reduction in error achieved by using $Y$. It measures how much the "best guesses" ($1/6$ and $1/2$) vary from the global average ($0.30$).

Calculation:

$$Var(E[X|Y]) = P(Y=0)(1/6 - 0.3)^2 + P(Y=1)(1/2 - 0.3)^2$$$$Var(E[X|Y]) = 0.6(1/6 - 1.8/6)^2 + 0.4(0.5 - 0.3)^2$$$$Var(E[X|Y]) = 0.6(-0.8/6)^2 + 0.4(0.2)^2 \approx \mathbf{0.0267}$$

C. The Total Uncertainty ($Var(X)$)

Summing them up:

$$0.1833 \text{ (Unexplained)} + 0.0267 \text{ (Explained)} = 0.21 \text{ (Total Variance)}$$

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

# Answer

1.
To show that the conditional distribution of $X$ given $Y=y$ is multivariate Gaussian with the specified parameters, we use the definition of conditional density:$$f_{X|Y}(x|y) = \frac{f_{X,Y}(x, y)}{f_Y(y)}$$This derivation relies on a technique called completing the square within the exponent of the joint density.

1. Partitioning the Quadratic Form

The exponent of the joint density contains the quadratic form:$$Q(x,y) = \begin{bmatrix} (x-\mu_X) \\ (y-\mu_Y) \end{bmatrix}^T \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}^{-1} \begin{bmatrix} (x-\mu_X) \\ (y-\mu_Y) \end{bmatrix}$$To evaluate this, we use the Schur Complement to invert the partitioned covariance matrix:$$\Sigma^{-1} = \begin{bmatrix} \mathbf{I} & -\Sigma_{XY}\Sigma_{YY}^{-1} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}^T \begin{bmatrix} (\Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX})^{-1} & \mathbf{0} \\ \mathbf{0} & \Sigma_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I} & -\Sigma_{XY}\Sigma_{YY}^{-1} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}$$Let $\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$ (the Schur complement of $\Sigma_{YY}$).

2. Decomposing the Exponent

Substitute this inverse back into $Q(x,y)$. After performing the matrix multiplication, the quadratic form splits into two parts:$$Q(x,y) = (x - \mu^*) ^T \Sigma_{X|Y}^{-1} (x - \mu^*) + (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y)$$where:$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$

3. Simplifying the Density Ratio

Now, substitute this decomposed $Q(x,y)$ into the exponential of the joint density $f_{X,Y}(x,y)$:$$f_{X,Y}(x,y) \propto \exp\left( -\frac{1}{2} (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*) \right) \cdot \exp\left( -\frac{1}{2} (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y) \right)$$When we divide by the marginal density $f_Y(y)$:$$f_Y(y) = \frac{1}{\sqrt{(2\pi)^m |\Sigma_{YY}|}} \exp\left( -\frac{1}{2} (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y) \right)$$The terms involving $(y - \mu_Y)$ in the exponent cancel out completely.

4. Handling the Determinants

Using the properties of partitioned matrices, the determinant of $\Sigma$ is:$$|\Sigma| = |\Sigma_{YY}| \cdot |\Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}| = |\Sigma_{YY}| \cdot |\Sigma_{X|Y}|$$When we divide the joint normalization factor by the marginal normalization factor:$$\frac{\frac{1}{\sqrt{(2\pi)^{n+m} |\Sigma|}}}{\frac{1}{\sqrt{(2\pi)^{m} |\Sigma_{YY}|}}} = \frac{1}{\sqrt{(2\pi)^{n} \frac{|\Sigma|}{|\Sigma_{YY}|}}} = \frac{1}{\sqrt{(2\pi)^{n} |\Sigma_{X|Y}|}}$$

Final Result

Putting it all together, we obtain:$$f_{X|Y}(x|y) = \frac{1}{\sqrt{(2\pi)^{n} |\Sigma_{X|Y}|}} \exp\left( -\frac{1}{2} (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*) \right)$$This is exactly the density of a Multivariate Gaussian $\mathcal{N}(\mu^*, \Sigma_{X|Y})$.

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



# Answers

1.
Joint Distribution of $(X, Y)$

A vector $(X, Y)$ has a joint multivariate normal distribution if every linear combination of $X$ and $Y$ is a univariate normal distribution. Since $X$ and $Q$ are independent normal variables, any linear combination of them is normal. Because $Y = X + Q$ is itself a linear combination of $X$ and $Q$, the vector $(X, Y)$ is Jointly Multivariate Normal.

To fully specify this distribution, we need the mean vector and the covariance matrix:

Mean vector: $\mu = \begin{bmatrix} E[X] \\ E[Y] \end{bmatrix}$

Covariance matrix: $\Sigma = \begin{bmatrix} Var(X) & Cov(X,Y) \\ Cov(Y,X) & Var(Y) \end{bmatrix}$

We determine the distribution of $Y$ by calculating its mean and variance based on the properties of $X$ and $Q$.

The Mean ($E[Y]$):

Using the linearity of expectation:$$E[Y] = E[X + Q] = E[X] + E[Q]$$Since $X \sim N(0,1)$ and $Q \sim N(0,4)$:$$E[Y] = 0 + 0 = \mathbf{0}$$

The Variance ($Var(Y)$):

Since $X$ and $Q$ are independent, the variance of their sum is the sum of their variances:$$Var(Y) = Var(X + Q) = Var(X) + Var(Q)$$$$Var(Y) = 1 + 4 = \mathbf{5}$$Therefore, $Y$ follows a normal distribution with mean 0 and variance 5: $Y \sim N(0, 5)$.

The covariance measures the joint variability of the hidden temperature and the observed power.

Using the definition of covariance and the fact that $Y = X + Q$:$$Cov(X, Y) = Cov(X, X + Q)$$Applying the property of bilinearity of covariance:$$Cov(X, X + Q) = Cov(X, X) + Cov(X, Q)$$$Cov(X, X)$: By definition, the covariance of a variable with itself is its variance: $Var(X) = \mathbf{1}$.

$Cov(X, Q)$: Since $X$ and $Q$ are independent, their covariance is $0$.$$Cov(X, Y) = 1 + 0 = \mathbf{1}$$

2.
In measure theory, the statement "$X$ is not $\sigma(Y)$-measurable" is the formal way of saying that the information provided by $Y$ is insufficient to uniquely determine the value of $X$.

To understand this from an information geometry and practical engineering perspective, let’s break it down using the "Signal + Noise" model you've established: $Y = X + Q$.

1. The Geometry of the "Pre-image"

Recall that for $X$ to be $\sigma(Y)$-measurable, its value must be constant on every set of outcomes where $Y$ is constant. Mathematically, knowing $Y=y$ would have to imply a single, unique $X=x$.

In this motor problem, look at the set of all outcomes $(\omega)$ that result in a power reading of $Y=5$:

It could be that $X=1$ (slight heat) and $Q=4$ (very hard material).

It could be that $X=5$ (dangerous overheat) and $Q=0$ (no material load).

It could be that $X=-1$ (cold motor) and $Q=6$ (extremely hard material).

All of these distinct physical realities are mapped to the same value $Y=5$. In geometry, the "isoline" for $Y=y$ is a diagonal line in the $(X, Q)$ plane. Because $X$ varies along that line, $X$ is not a function of $Y$.

2. The Information "Blur"

From a $\sigma$-algebra perspective, $\sigma(Y)$ is a "coarser" collection of information than the full sample space.

The sensor $Y$ partitions the reality into "slices."

Within each slice (where $Y$ is fixed), there is still a distribution of possible $X$ values.

If you try to "view" $X$ through the lens of $\sigma(Y)$, you don't see a point; you see a spread (a conditional variance). Because that spread is greater than zero, $X$ cannot be $\sigma(Y)$-measurable.

3. Practical Engineering Reality: The "Ambiguity"

If an engineer sees the meter spike to $Y=5$, they face a System Identifiability problem. There are two competing "theories" for why the power is high:

The Thermal Theory: The motor is failing/overheating ($X \uparrow$).

The Mechanical Theory: The CNC bit just hit a hard knot in the wood or a harder patch of steel ($Q \uparrow$).

Because $Y$ only measures the sum of these effects, the sensor is "blind" to the source of the energy. In practical terms:

If the engineer assumes $X$ is high and shuts down the motor, but it was actually just hard material, the factory suffers unnecessary downtime (False Positive).

If the engineer assumes $Q$ is high and keeps cutting, but the motor was actually overheating, the spindle might melt (False Negative).

4. How Conditional Expectation $E[X|Y]$ helps

Since the engineer cannot know $X$ for certain, the best they can do is use the conditional expectation. As we saw in the Gaussian derivation:$$E[X|Y] = \frac{Cov(X,Y)}{Var(Y)}(Y - \mu_Y)$$Substituting your specific values ($Cov=1, Var=5$):$$E[X|Y] = \frac{1}{5}Y = 0.2Y$$

If $Y=5$, the "best estimate" is $X=1$.

Why not $X=5$? Because the math (Information Geometry) recognizes that $Q$ has a much higher variance ($Var=4$) than $X$ ($Var=1$). It is statistically "cheaper" to explain a power spike by blaming the noisy mechanical load than by assuming the temperature has swung to an extreme. This is known as Shrinkage—the estimate "shrinks" away from the noisy observation toward the mean.

3.
To find the "best estimate" of the hidden temperature $X$ given the observed power $Y$, we apply the linear projection formula for joint multivariate normal distributions.

1. Computing $E[X|Y]$

Using the parameters we derived in the previous section:

$\mu_X = 0$

$\mu_Y = 0$

$Cov(X, Y) = 1$

$Var(Y) = 5$

Substitute these into the conditional expectation formula:

$$E[X|Y=y] = \mu_X + \frac{Cov(X, Y)}{Var(Y)}(y - \mu_Y)$$$$E[X|Y=y] = 0 + \frac{1}{5}(y - 0)$$$$E[X|Y=y] = \frac{1}{5}y = 0.2y$$In terms of the random variable $E[X|Y]$, we write:$$E[X|Y] = \frac{1}{5}Y$$

2. The "Signal-to-Total-Variance" Ratio

In the signal processing context of this problem, $X$ is our Signal and $Y$ is our Total Observation (Signal + Noise). Let's look at the coefficient $\frac{1}{5}$ more closely:

Since $Y = X + Q$ and $X, Q$ are independent, we know:

Signal Variance: $Var(X) = 1$

Total Variance: $Var(Y) = Var(X) + Var(Q) = 1 + 4 = 5$

Therefore, the coefficient can be expressed as:$$\frac{Cov(X, Y)}{Var(Y)} = \frac{Var(X)}{Var(X) + Var(Q)} = \frac{\text{Signal Variance}}{\text{Total Variance}}$$

3. Physical Interpretation: The "Trust" Factor

This ratio ($\frac{1}{5}$ or $20\%$) represents how much the bank or engineer should trust the sensor reading $Y$ as a reflection of the true state $X$.

If Noise is Low ($Var(Q) \approx 0$): The ratio approaches $1$. The estimate $E[X|Y] \approx Y$. We trust the sensor completely.

If Noise is High ($Var(Q) \to \infty$): The ratio approaches $0$. The estimate $E[X|Y] \approx 0$ (the mean). We ignore the sensor because it's too noisy and stick to our "prior" knowledge.
